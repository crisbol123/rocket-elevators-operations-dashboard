## 1. Load and document dataset structure
Load `installed.json` into a DataFrame using `pd.read_json`. Use `.shape`, `.dtypes`, and `.head()` to document the number of records, column names, and data types.

```python
import pandas as pd

installed_df = pd.read_json("../data/installed.json")

print("Shape:", installed_df.shape)
```

## 2. Device type distribution
Identify which device types appear in the dataset and how frequently using `value_counts()`. Visualize the counts as a horizontal bar chart.

```python
import matplotlib.pyplot as plt

device_counts = installed_df["Device Type"].value_counts()

print(device_counts)

fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(device_counts.index[::-1], device_counts.values[::-1], color="steelblue", edgecolor="white")

ax.set_xlabel("Number of Devices")
ax.set_ylabel("Device Type")
ax.set_title("Installed Elevating Devices by Type (Ontario)")

plt.tight_layout()
plt.show()
```

## 4. Comparison with license.csv

installed.json describes the physical device (type and current device status). license.csv describes the regulatory license (holder, expiry date, and license status). The best join key is the device identifier: Elevating devices number in installed.json matches ElevatingDevicesNumber in license.csv. Location can be used as a secondary match. DeviceStatus and LICENSESTATUS are different concepts, so they should be analyzed separately when the datasets are joined.
