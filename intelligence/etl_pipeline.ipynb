## AND-102 Task 5: Merge 1, License + Installed — Part 1

```python
import pandas as pd

license_df = pd.read_csv("../data/license.csv")


KEEP = {"ACTIVE", "PENDING_RENEWAL"}

license_df = license_df[license_df["LICENSESTATUS"].isin(KEEP)].reset_index(drop=True)
print(license_df.shape)
license_df.head()
```

```python
installed_df = pd.read_json("../data/installed.json")

print(installed_df.shape)
installed_df.head()
```

```python
# Merge 1 start — row counts before joining
print(f'license_df:   {len(license_df):,} rows')
print(f'installed_df: {len(installed_df):,} rows')
```

## AND-102 Task 5: Merge 1, License + Installed — Part 2

```python
# Common key: 'ElevatingDevicesNumber' (license) <-> 'Elevating devices number' (installed)
rows_license   = len(license_df)
rows_installed = len(installed_df)

merged_df = license_df.merge(
    installed_df,
    left_on='ElevatingDevicesNumber',
    right_on='Elevating devices number',
    how='inner'
)

print(f'license_df rows:   {rows_license:,}')
print(f'installed_df rows: {rows_installed:,}')
print(f'merged_df rows:    {len(merged_df):,}')

merged_df.head()
```

## AND-102 Task 5: Merge 1, License + Installed — Part 3

```python
# Step 1: Compare location columns — identical vs different
identical_mask = merged_df['LocationoftheElevatingDevice'] == merged_df['Location of Device']

print(f'Identical location values: {identical_mask.sum():,}')
print(f'Different location values:  {(~identical_mask).sum():,}')
print(f'Match rate: {identical_mask.mean():.2%}')
```

```python
# Step 2: Inspect rows where location values disagree
discrepancies = merged_df[~identical_mask][['ElevatingDevicesNumber',
                                               'LocationoftheElevatingDevice',
                                               'Location of Device']].reset_index(drop=True)
discrepancies.head(10)
```

```python
# Step 3: Extract postal codes from each location column
# Canadian postal code format: A1A 1A1
import re

postal_pattern = r'[A-Z]\d[A-Z]\s\d[A-Z]\d'

merged_df = merged_df.copy()
merged_df['postal_license']   = merged_df['LocationoftheElevatingDevice'].str.extract(f'({postal_pattern})')
merged_df['postal_installed'] = merged_df['Location of Device'].str.extract(f'({postal_pattern})')

disc = merged_df[~identical_mask]
postal_match = disc['postal_license'] == disc['postal_installed']

print(f'Among {len(disc)} discrepant rows:')
print(f'  Postal codes match (abbreviation/city-name variant): {postal_match.sum()}')
print(f'  Postal codes differ (genuine address mismatch):       {(~postal_match).sum()}')
```

```python
rows_before = len(merged_df)

postal_match_all = merged_df['postal_license'] == merged_df['postal_installed']
merged_df = merged_df[postal_match_all].reset_index(drop=True)

print(f'Rows before postal filter: {rows_before:,}')
print(f'Rows after postal filter:  {len(merged_df):,}')
print(f'Rows removed:              {rows_before - len(merged_df):,}')

merged_df.head()
```

# Identify problematic categories
-Freight Elevator-P' and 'Freight Elevator-E': cryptic suffixed variants of Freight Elevator

-Material Lift - ATD', 'Special Installation', 'Power Type Manlift': single-occurrence values


```python
# Clean 'Device Type':

#   - Merge 'Freight Elevator-P' and 'Freight Elevator-E' into 'Freight Elevator'
#     Reason: suffixes -P/-E are undocumented subtypes with only 28 total rows;
#     consolidating avoids misleading sparse categories in analysis.


#   - Map the three single-occurrence values to 'Specialized Equipment'
#     Reason: 1-row categories have no statistical weight and clutter groupby operations.
#   Result: 11 categories → 8 clean, meaningful categories.

remap = {
    'Freight Elevator-P': 'Freight Elevator',
    'Freight Elevator-E': 'Freight Elevator',
    'Material Lift - ATD': 'Specialized Equipment',
    'Special Installation': 'Specialized Equipment',
    'Power Type Manlift': 'Specialized Equipment',
}

merged_df['Device Type'] = merged_df['Device Type'].replace(remap)

print(f'Categories before: 11')
print(f'Categories after:  {merged_df["Device Type"].nunique()}')
print()
print(merged_df['Device Type'].value_counts(dropna=False))
```

```python
# Merge 1 end — final shape and explanation
print(f'Merge 1 final shape: {merged_df.shape}')
print()
print('Rows lost:')
print('  - 3,639 from installed_df: devices with no active or pending license')
print('    excluded by the inner join — outside the scope of this pipeline.')
print('  - 143 from the postal code filter, broken into two causes:')
print('      129 rows where postal codes genuinely differed between the license')
print('      and installed location columns — real address mismatches.')
print('      14 rows where the regex could not extract a postal code from one')
print('      or both columns (null/malformed address). In pandas, NaN == NaN')
print('      evaluates to False, so these rows also fail the filter.')
```

```python
# Load altered.json
altered_df = pd.read_json('../data/altered.json')

print(altered_df.shape)
print(altered_df.columns.tolist())
altered_df.head()
```

```python
rows_before = len(merged_df)

merged_df = merged_df.merge(
    altered_df,
    left_on='ElevatingDevicesNumber',
    right_on='Elevating Devices Number',
    how='left'
)

print(f'Rows before merge: {rows_before:,}')
print(f'Rows after merge:  {len(merged_df):,}')
print(f'Rows gained:       {len(merged_df) - rows_before:,}')
print(f'Elevators with no alteration record: {merged_df["Elevating Devices Number"].isna().sum():,}')
```

```python
# Merge 2 end — final shape and explanation
print(f'Merge 2 final shape: {merged_df.shape}')
print()
print('Rows gained:')
print('  - altered.json has one row per alteration event.')
print('    Elevators altered multiple times generate multiple rows in the join.')
print('  - Elevators never altered are kept with NaN in alteration columns (left join).')
```

## AND-102 Task 5: Merge 3, License + Installed + Altered + Inspections — Part 1

```python
# Merge 3 start — row count entering this merge
print(f'merged_df entering Merge 3: {len(merged_df):,} rows')
```

```python
inspection_df = pd.read_csv('../data/inspection.csv')

print(inspection_df.shape)
print(inspection_df.columns.tolist())
inspection_df.head()
```

```python
# Can one elevator have multiple inspections?
per_elevator = inspection_df.groupby('ElevatingDevicesNumber')['InspectionNumber'].count()

print('Inspections per elevator:')
print(f'  Min:  {per_elevator.min()}')
print(f'  Max:  {per_elevator.max()}')
print(f'  Mean: {per_elevator.mean():.2f}')
print(f'  Elevators with >1 inspection: {(per_elevator > 1).sum():,} of {len(per_elevator):,}')
print()

# Can one inspection cover multiple elevators?
per_inspection = inspection_df.groupby('InspectionNumber')['ElevatingDevicesNumber'].count()

print('Elevators per inspection:')
print(f'  Min:  {per_inspection.min()}')
print(f'  Max:  {per_inspection.max()}')
print(f'  Mean: {per_inspection.mean():.2f}')
print(f'  Inspections covering >1 elevator: {(per_inspection > 1).sum():,} of {len(per_inspection):,}')
```

**Findings:**

One elevator can absolutely have multiple inspections — on average 3.5 per elevator, with some reaching as many as 24. That makes sense: inspections happen periodically over the life of a device, so repeat records are expected, not errors.

The other direction is clean: every InspectionNumber maps to exactly one elevator. No inspection covers more than one device. The relationship is strictly one-to-many (elevator → inspections).

This means the merge will expand the row count, one row per inspection event per elevator. Elevators with no inspection record will keep a single row with NaN in the inspection columns.

## AND-102 Task 5: Merge 3, License + Installed + Altered + Inspections — Part 2

**Approach: aggregate inspection history per elevator**

A straight left join would expand rows to ~180K — one per inspection event — which breaks the one-row-per-elevator structure needed for the dashboard. Keeping only the most recent inspection is simple but throws away history.

Aggregating before the join gives the best of both: one row per elevator, but richer data. For each elevator we compute total inspections, date range, last outcome, and pass rate. Pass rate is defined as the proportion of outcomes that are unambiguously positive: `Passed`, `Passed Major`, `Passed Sub`, `All Orders Resolved`, `Complete`, `Inspection Complete`, or `Complete Enforcement`.

```python
inspection_df['Latest_INSPECTION_Date'] = pd.to_datetime(
    inspection_df['Latest_INSPECTION_Date'], format='%m/%d/%Y'
)
inspection_df['Earliest_INSPECTION_Date'] = pd.to_datetime(
    inspection_df['Earliest_INSPECTION_Date'], format='%m/%d/%Y'
)

PASS_OUTCOMES = {
    'Passed', 'Passed Major', 'Passed Sub',
    'All Orders Resolved', 'Complete',
    'Inspection Complete', 'Complete Enforcement'
}

inspection_df['is_passed'] = inspection_df['InspectionOutcome'].isin(PASS_OUTCOMES)

# Most recent outcome per elevator (for last_outcome column)
last_outcome = (
    inspection_df
    .sort_values('Latest_INSPECTION_Date', ascending=False)
    .drop_duplicates(subset='ElevatingDevicesNumber', keep='first')
    [['ElevatingDevicesNumber', 'InspectionOutcome', 'InspectionType']]
    .rename(columns={
        'InspectionOutcome': 'last_inspection_outcome',
        'InspectionType': 'last_inspection_type'
    })
)

# Aggregate metrics per elevator
inspection_agg = (
    inspection_df
    .groupby('ElevatingDevicesNumber')
    .agg(
        total_inspections=('InspectionNumber', 'count'),
        first_inspection_date=('Earliest_INSPECTION_Date', 'min'),
        last_inspection_date=('Latest_INSPECTION_Date', 'max'),
        pct_passed=('is_passed', 'mean')
    )
    .reset_index()
)

inspection_agg = inspection_agg.merge(last_outcome, on='ElevatingDevicesNumber', how='left')
inspection_agg['pct_passed'] = inspection_agg['pct_passed'].round(2)

print(inspection_agg.shape)
inspection_agg.head()
```

```python
rows_before = len(merged_df)

merged_df = merged_df.merge(
    inspection_agg,
    on='ElevatingDevicesNumber',
    how='left'
)

print(f'Rows before merge: {rows_before:,}')
print(f'Rows after merge:  {len(merged_df):,}')
print(f'Rows gained/lost:  {len(merged_df) - rows_before:,}')
print(f'Elevators with no inspection record: {merged_df["total_inspections"].isna().sum():,}')
```

```python
# Merge 3 end — final shape and explanation
print(f'Merge 3 final shape: {merged_df.shape}')
print()
print('Rows gained/lost: 0')
print('  - inspection_df (143,181 rows) was aggregated to one row per elevator')
print('    before the join, so the left join does not expand rows.')
print('  - Elevators with no inspection history are kept with NaN (left join).')
```
